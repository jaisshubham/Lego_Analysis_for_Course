### Aggregate Data with the Python .agg() Function

Let's work out the number of different themes shipped by year. This means we have to count the number of unique theme_ids per calendar year.

```python
sets
```

```python
theme_by_year=sets.groupby('year').agg({"theme_id":pd.Series.nunique})
print(theme_by_year.head())
theme_by_year.tail()
```

```python
theme_by_year.rename(columns={'theme_id':'nr_theme'},inplace=True)
print(theme_by_year.head())
theme_by_year.tail()
```

**Challenge**: Plot the number of themes released by year on a line chart. Only include the full calendar years (i.e., exclude 2020 and 2021).

```python
plt.plot(theme_by_year.index[:-2],theme_by_year.nr_theme[:-2],linewidth=2)
```

### Line Charts with Two Seperate Axes

```python
# It looks terriable, Difficult to analyze.
plt.plot(theme_by_year.index[:-2],theme_by_year.nr_theme[:-2],linewidth=2)
plt.plot(sets_year_data.index[:-2],sets_year_data['set_num'][:-2])
```

###The problem is that the "number of themes" and the "number of sets" have very different scales. The theme number ranges between 0 and 90, while the number of sets ranges between 0 and 900.
**<h1>So we need two separate axes</h1>**

### Now, Two separate axis plot.

```python

ax1=plt.gca() # get the axis.
ax2=ax1.twinx() #create another axis that share the same x axis.

# we here also add styling
ax2.plot(sets_year_data.index[:-2],sets_year_data.set_num[:-2],linewidth=2,color='g')
ax1.plot(theme_by_year.index[:-2],theme_by_year.nr_theme[:-2],linewidth=2,color='b')

# assigning label to the axes and
ax1.set_xlabel('Years',fontsize=14)
ax1.set_ylabel('No of sets',color='b',fontsize=14)
ax2.set_ylabel("No of themes",color='green',fontsize=14)
```

### Merging (i.e., Combining) DataFrames based on a Key


```python
# here we are combining two  data frame by .merge() method
# but we should have same column name in both the database.
set_theme_count=sets['theme_id'].value_counts()
themes_data=pd.read_csv("/content/data/themes.csv")
```

###**let's give set_theme_count's column a name..**

```python
set_theme_count=pd.DataFrame({'id':set_theme_count.index,'set_count':set_theme_count.values})
set_theme_count.head()
```

```python
# Now merge these two DataFrame along one column having name as 'id'
merged_df=pd.merge(set_theme_count,themes_data,on='id')
merged_df.head()
```

<h3>Now ploting Bar graph. </h3>

```python
plt.bar(merged_df.name[:10],merged_df.set_count[:10])
```
